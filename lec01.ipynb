```python
from numpy import average
import matplotlib.pyplot as plt
import random
import math
import statistics

# a single experiment
def single_trial():
  num_rolls = 0
  while True:
    num_rolls = num_rolls +1
    die_roll = random.randint(1,6) #fair die roll
    if die_roll == 6:
      break
  return num_rolls

# perform N trials
N = 10000
rolls = []
trial = [i+1 for i in range(N)]
for i in range(N):
  num_rolls = single_trial()
  rolls.append(num_rolls)

plt.title('Average '+str(average(rolls)))
plt.ylabel('Number of rolls')
plt.bar(trial, rolls)
```

```python
from numpy import average
import matplotlib.pyplot as plt
import random
import math
import statistics

# a single experiment
def single_trial():
  num_rolls = 0
  while True:
    num_rolls = num_rolls +1
    die_roll = random.randint(1,6) #fair die roll
    if die_roll % 2: # restart the experiment
      num_rolls = 0
    if die_roll == 6:
      break
  return num_rolls

# perform N trials
N = 10
rolls = []
trial = [i+1 for i in range(N)]
for i in range(N):
  num_rolls = single_trial()
  rolls.append(num_rolls)

plt.title('Average '+str(average(rolls)))
plt.ylabel('Number of rolls')
plt.bar(trial, rolls)
```
